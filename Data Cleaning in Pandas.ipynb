```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\marth\Downloads\Customer Call List.xlsx")
```

```python
#names as unwanted characters, 
#phone numbers needs cleaning, 
#address - some have street/+othr locations, 
#paying customers - y n yes no, do not contact - same y n ..., 
#not useful columns - get rid of it
```

```python
#remove duplicates
df = df.drop_duplicates()
```

```python
#drop unnecessary column
df = df.drop(columns='Not_Useful_Column')
```

```python
df
```

```python
#fix last names
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
#OR
df["Last_Name"].str.strip("123._/")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
```

```python
df
```

## try this with alternate method, repeat above steps except ltrim, rtrim

```python
#cleaning up phone numbers
df["Phone_Number"]=df["Phone_Number"].str.replace('[^a-xA-Z0-9]','')
```

```python
#using lambda

#df["Phone_Number"].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:10]
#TypeError: 'float' object is not subscriptable

#str(df["Phone_Number"]) - does not work well

#df["Phone_Number"]=df["Phone_Number"].apply(lambda x:str(x))
#df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
```

```python
df
```

```python
df[["Street Address","State","Zip Code"]]=df["Address"].str.split(',',2,expand=True)
```

```python
df
```

```python
df["Paying Customer"]=df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"]=df["Paying Customer"].str.replace('No','N')
```

```python
df
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
df=df.replace('N/a','')
df=df.replace('NaN','')#--this will not work so do fillna
df
```

```python
df.fillna('')
```

```python
#lets remove those which do not call is yes and missing phone number
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df
```

```python
#df["Do_Not_Contact"].str.replace('','N') - didnt work, move on
#lets remove those which do not call is yes and missing phone number
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
df
#we could have also done below to drop null values-
df.dropna(subset="Phone_Number"), inplace= True)
```
