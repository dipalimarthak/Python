# Exploratory Data Analysis in Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r"C:\Users\marth\Downloads\world_population EDA.csv")
df
```

```python
pd.set_option('display.float_format',  lambda x: '%.2f' % x)
df
```

```python
df.info()
```

```python
#observe we started losing values since 2022
df.describe()
```

```python
df.isnull().sum()
#we can choose to replace these nulls with median values
#or we can choose to completely delete these values
```

```python
df.nunique()
```

```python
#lets find top range of countries

df.sort_values(by="2022 Population", ascending=False).head()
```

```python
df.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
#lets find correlation(usually happens in numeric values)
#1, 0.99 --- highly correlated
df.corr()
```

```python
sns.heatmap(df.corr(), annot = True)
plt.rcParams['figure.figsize'] = (20,20)
plt.show()
```

```python
#let's group on the continent
#and lets see if certain continents have grown faster
df.groupby('Continent').mean().sort_values(by="2022 Population",ascending=False)
```

```python
df[df['Continent'].str.contains('Oceania')]
```

```python
df2 = df.groupby('Continent').mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
df2.plot()
```

```python
df3 = df2.transpose()
```

```python
df3.plot()
```

```python
df.columns
```

```python
df2 = df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
df2.plot()
```

```python
df3.plot() #this is backward plotting
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
df2.plot()
```

```python
df3 = df2.transpose()
df3.plot()
```
